## Final Listing

```python
for t in times:
    formatted_strings = t.dt.strftime('%Y%m%dT%H%M')
    print( formatted_strings.data )
```

```python
t0 = datetime(2020,6,1)
t1 = datetime(2020,6,30)
dt = timedelta(days = 1)
t = t0
while t<=t1:
    print(t.strftime('%Y%m%d'))
    t += dt
```
